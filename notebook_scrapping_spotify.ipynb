```python
from google_play_scraper import reviews, Sort
app_ID = 'com.spotify.music'

result, _ = reviews(
    app_ID,
    lang='id',
    country='id',
    sort = Sort.NEWEST,
    count= 11000
)

print(f"Jumlah data yang diambil: {len(result)}")
```

```python
import pandas as pd
df_reviews = pd.DataFrame(result)
df_2024 = df_reviews[df_reviews['at'].dt.year == 2024]
df_reviews.head()
```

```python
df_reviews.to_csv('reviews_spotify.csv', index =False)

from google.colab import files
files.download('reviews_spotify.csv')
```
